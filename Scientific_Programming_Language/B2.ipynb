```python
import math
```

```python
def Euclidean (point_1, point_2):
    distance = float( math.sqrt((point_1[0]-point_2[0])**2 + (point_1[1]-point_2[1])**2) )
    return distance
```

```python
def check_distence_with_O (point_1, point_2):
    return point_1 if Euclidean(point_1, [0,0]) < Euclidean(point_2, [0,0]) else point_2
```

```python
def perimeter_OAB(point_1, point_2):
    return Euclidean(point_1, [0,0]) + Euclidean(point_2, [0,0]) + Euclidean(point_1, point_2)
```

```python
point_A = list(map(int, input("nhập lần lượt x và y của A: ").split()))
point_B = list(map(int, input("nhập lần lượt x và y của B: ").split()))
print(point_A,point_B, end="\n")
```

```python
print(f"distance of A and B: {Euclidean(point_1= point_A,point_2= point_B):.2f}")
print(f"perimeter OAB = {perimeter_OAB(point_1=point_A,point_2= point_B):.2f}")
result = check_distence_with_O(point_A, point_B)
if result == point_A:
    print(f"A is closer to O")
else:
    print(f"B is closer to O")
```

```python
a, b, c, d, e = 5, 4, 3, 2, 7
def max_min(*tuple):
    max = -10e10
    min = 10e10
    for i in tuple:
        if max < i:
            max = i
        if min > i:
            min = i
    return max, min
max_min(a, b, c, d, e)
```
