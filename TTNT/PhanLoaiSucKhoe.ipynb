```python
import category_encoders as ce
encoder = ce.OrdinalEncoder(cols = ['Mau da', 'Chieu cao', 'Can nang', 'The thao', 'Suc khoe'])
df = encoder.fit_transform(df)
df.head(14)
```

```python
inputs=df.drop('Suc khoe',axis='columns')
target=df['Suc khoe']
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
model.score(inputs,target)
```

```python
# Nhập dữ liệu mới để dự đoán
mauda = float(input("Enter mau da: "))
chieucao = float(input("Enter chieu cao: "))
cannang = float(input("Enter can nang: "))
thethao = float(input("Enter the thao: "))
# Tạo một điểm dữ liệu mới từ thông tin người dùng nhập vào
new_data_point = [mauda,chieucao, cannang, thethao]

result = model.predict([new_data_point])
# In ra kết quả dự đoán
print("Predicted fruit:", result[0])
```
